```python
import seaborn

df = seaborn.load_dataset("titanic")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
df["fare"].plot(kind="kde")
```

```python
# sample size = 50
sample = []
for i in range(100):
  sample.append(df["fare"].sample(50).values.tolist())
```

```python
samples = np.array(sample)
```

```python
sampling_means = samples.mean(axis=1) # mean of each row
```

```python
seaborn.kdeplot(sampling_means)
```

```python
sampling_means.mean()
```

```python
sampling_means.std()/np.sqrt(50)
```

```python
lower_limit = sampling_means.mean() - 2 * sampling_means.std()/np.sqrt(50)
upper_limit = sampling_means.mean() + 2 * sampling_means.std()/np.sqrt(50)
```
